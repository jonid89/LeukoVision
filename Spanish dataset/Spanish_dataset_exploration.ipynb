```python
# Import libraries
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from glob import glob
import cv2
```

```python
base_path = "G:/My Drive/DataScientest - Blood cells Project/Spanish Dataset"

subfolders = os.listdir(base_path)
imgs_per_cell = {}

for subfolder in subfolders:
    subfolder_path = os.path.join(base_path, subfolder)
    images = glob(os.path.join(subfolder_path, "*.jpg"))
    imgs_per_cell[subfolder] =  len(images)
```

```python
fig = plt.figure(figsize=(10,5))
plt.bar(x= imgs_per_cell.keys(), height=imgs_per_cell.values())
plt.ylabel('Number of cell images')
plt.xlabel('Blood cell type');
```
